```python
tips = sns.load_dataset('tips')
```

```python
px.bar(tips, x = 'day' , y = 'tip' , color = 'day' , title ='Tips by Day')
```

```python
px.bar(tips, x = 'day' , y = 'total_bill' , color = 'day' , title ='Total Bill by Day')
```

```python
px.scatter(tips,x = 'total_bill', y ='tip', title ='Tips by Day', color ='day',size='size')
```

```python
px.histogram(tips, x='total_bill',title='Total Bill Distribution',nbins=50)
```

```python
px.pie(tips,names = "sex")
```

```python
sns.barplot(data=tips, x='day', y='total_bill')
```
